```python
from dotenv import load_dotenv
import os
load_dotenv()

from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(
    model='text-embedding-3-large',
    # model="text-embedding-3-small",
    api_key=os.getenv("OPENAI_KEY"),
    base_url=os.getenv("BASE_URL"),
)

INPUT_FILE_ADVANCE = "../inputs/menu_advanceAi.json"
DB_DIR_ADVANCE = "../ChromaDB/db_menu_advance"

INPUT_FILE_TERSE = "../inputs/menu_terseAi.json"
DB_DIR_TERSE = "../ChromaDB/db_menu_terse"

import sys
from pathlib import Path

sys.path.append(str(Path.cwd().parent))
```

```python
from libAgent.jsonFileToDocument import jsonFileToDocuments

chunckAdvance = jsonFileToDocuments(INPUT_FILE_ADVANCE)
chunckTerse = jsonFileToDocuments(INPUT_FILE_TERSE)
```

```python
from langchain_chroma import Chroma

vStore = Chroma(
    embedding_function=embeddings,
    persist_directory=DB_DIR_ADVANCE
)

vStore.add_documents(chunckAdvance)
```

```python
from langchain_chroma import Chroma

vStore = Chroma(
    embedding_function=embeddings,
    persist_directory=DB_DIR_TERSE
)

vStore.add_documents(chunckTerse)
```
